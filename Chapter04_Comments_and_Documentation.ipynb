Python bỏ qua các chú thích, do đó nó sẽ không thực thi các đoạn mã bên trong đó, hoặc ném ra lỗi cú pháp (syntax errors) cho các câu văn tiếng Anh (hoặc tiếng Việt) thông thường.

# Phần 4.2: Truy xuất chuỗi tài liệu thông qua lập trình (Programmatically accessing docstrings)

Khác với các chú thích thông thường, chuỗi tài liệu (docstrings) được lưu trữ dưới dạng một thuộc tính (attribute) của hàm mà chúng mô tả, nghĩa là bạn có thể truy xuất chúng thông qua lập trình.

```python
def func():
  """Đây là một hàm không làm gì cả"""
  return
```

Docstring có thể được truy cập thông qua thuộc tính __doc__:

```python
print(func.__doc__)
```

```python
help(func)
```

function.doc chỉ trả về đúng chuỗi docstring dưới dạng một biến chuỗi (string), trong khi hàm help cung cấp thông tin tổng quan về một hàm, bao gồm cả docstring của nó. Đây là một ví dụ hữu ích hơn:

```python
def greet(name, greeting = "Hello"):
  '''In ra lời chào cho người dùng 'name'
  Tham số tùy chọn 'greeting' có thể thay đổi cách họ được chào đón.'''
  print("{} {}".format(greeting, name))

help(greet)
```

**Ưu điểm của docstrings so với chú thích thông thường**

Việc không đặt docstring hoặc chỉ dùng chú thích thông thường trong một hàm khiến hàm đó trở nên kém hữu ích hơn rất nhiều.

```python
def greet(name, greeting="Hello"):
    # In ra lời chào cho người dùng `name`
    # Tham số tùy chọn `greeting` có thể thay đổi cách họ được chào đón.
    print("{} {}".format(greeting, name))

print(greet.__doc__)
```

# Phần 4.3: Viết tài liệu sử dụng docstrings (Write documentation using docstrings)

Docstring là một dạng chú thích nhiều dòng được dùng để viết tài liệu cho các mô-đun, lớp, hàm và phương thức. Nó bắt buộc phải là câu lệnh đầu tiên bên trong thành phần mà nó mô tả.

```python
def hello(name):
    """Chào ai đó.
    In ra một lời chào ("Hello") cho người có tên được cung cấp.
    """
    print("Hello "+name)

class Greeter:
    """Một đối tượng dùng để chào hỏi mọi người.
    Nó chứa nhiều hàm chào hỏi cho một vài ngôn ngữ
    và các thời điểm trong ngày.
    """
```

Giá trị của docstring có thể được truy cập bên trong chương trình và - ví dụ - được sử dụng bởi lệnh help.

**Các quy ước cú pháp (Syntax conventions)**

**PEP 257**
PEP 257 định nghĩa một tiêu chuẩn cú pháp cho các chú thích docstring. Về cơ bản, nó cho phép hai loại:

  * **Docstrings Một dòng (One-line Docstrings):**
  
Theo PEP 257, chúng nên được dùng cho các hàm ngắn và đơn giản. Mọi thứ được đặt trên một dòng, ví dụ:

```python
def hello():
    """Say hello to your friends."""
    print("Hello my friends!")
```

Docstring này nên kết thúc bằng một dấu chấm, và động từ nên ở thể mệnh lệnh (imperative form).

  * **Docstrings Nhiều dòng (Multi-line Docstrings):**
  
Docstring nhiều dòng nên được sử dụng cho các hàm, mô-đun hoặc lớp dài hơn, phức tạp hơn.  

```python
def hello(name, language="en"):
    """Say hello to a person.

    Arguments:
    name: the name of the person
    language: the language in which the person should be greeted
    """
    print(greeting[language]+" "+name)
```

**Sphinx**
Sphinx là một công cụ dùng để tạo tài liệu định dạng HTML cho các dự án Python dựa trên docstrings. Ngôn ngữ đánh dấu (markup language) mà nó sử dụng là reStructuredText. Họ định nghĩa các tiêu chuẩn riêng cho việc viết tài liệu. Định dạng Sphinx được sử dụng phổ biến, ví dụ như trong IDE pyCharm.

Một hàm sẽ được ghi tài liệu như sau bằng định dạng Sphinx/reStructuredText:

```python
def hello(name, language="en"):
    """Say hello to a person.

    :param name: the name of the person
    :type name: str
    :param language: the language in which the person should be greeted
    :type language: str
    :return: a number
    :rtype: int
    """
    print(greeting[language]+" "+name)
    return 4
```
