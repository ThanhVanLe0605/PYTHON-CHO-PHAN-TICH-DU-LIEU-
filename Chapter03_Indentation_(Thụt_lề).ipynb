# Section 3.1: Simple example



Trong Python, Guido van Rossum đã dựa trên việc thụt lề (indentation) để nhóm các câu lệnh (statements). Lý do cho điều này được giải thích trong phần đầu tiên của "Design and History Python FAQ". Dấu hai chấm, :, được sử dụng để khai báo một khối lệnh (code block) có thụt lề, ví dụ như sau:

```python
class ExampleClass:
  # Mọi hàm thuộc về một lớp (class) đều phải được thụt lề như nhau
  def __init__(self):
    name = "example"

  def someFunction(self, a):
    # Lưu ý mọi thứ thuộc về một hàm đều phải thụt lề
    if a > 5:
      return True
    else:
      return False
# Nếu một hàm không được thụt lề cùng mức, nó sẽ không được coi là một phần của lớp cha (parent class)
def separateFunction(b):
  for i in b:
    if i == 1:
      return True
  return False
separateFunction([2, 3, 5, 6, 1])
```

**Dấu cách (Spaces) hay Tab?**

Mức thụt lề được khuyến nghị là 4 dấu cách, nhưng bạn có thể dùng tab hoặc dấu cách miễn là phải nhất quán. Tuyệt đối không trộn lẫn tab và dấu cách trong Python vì điều này sẽ gây ra lỗi (error) trong Python 3 và có thể gây lỗi trong Python 2.

# Section 3.2: How indentation is Parsed

Khoảng trắng (Whitespace) được xử lý bởi bộ phân tích từ vựng (lexical analyzer) trước khi được phân tích cú pháp (parsed).

Bộ phân tích từ vựng sử dụng một ngăn xếp (stack) để lưu trữ các cấp độ thụt lề. Ban đầu, ngăn xếp chỉ chứa giá trị 0, tương ứng với vị trí ngoài cùng bên trái. Bất cứ khi nào một khối lệnh lồng nhau (nested block) bắt đầu, cấp độ thụt lề mới sẽ được đẩy (pushed) vào ngăn xếp, và một thẻ "INDENT" (token) sẽ được chèn vào luồng thẻ (token stream) để chuyển tới trình phân tích cú pháp (parser). Sẽ không bao giờ có nhiều hơn một thẻ "INDENT" nằm liền nhau (nếu có sẽ gây lỗi IndentationError).

Khi gặp một dòng có cấp độ thụt lề nhỏ hơn, các giá trị sẽ được lấy ra (popped) khỏi ngăn xếp cho đến khi giá trị trên cùng bằng với cấp độ thụt lề mới (nếu không tìm thấy giá trị nào khớp, lỗi cú pháp sẽ xảy ra). Với mỗi giá trị được lấy ra, một thẻ "DEDENT" sẽ được tạo ra. Rõ ràng, có thể có nhiều thẻ "DEDENT" nằm liền kề nhau.

Bộ phân tích từ vựng sẽ bỏ qua các dòng trống (chỉ chứa khoảng trắng hoặc có thể là chú thích), và sẽ không bao giờ tạo ra thẻ "INDENT" hoặc "DEDENT" cho những dòng này.

Ở cuối mã nguồn, các thẻ "DEDENT" sẽ được tạo ra cho mỗi cấp độ thụt lề còn lại trong ngăn xếp, cho đến khi chỉ còn lại giá trị 0.

```python
if foo:
    if bar:
        x = 42
else:
    print foo
```

```python
# Đoạn mã trên được phân tích thành (kèm trạng thái ngăn xếp):

# <if> <foo> <:>  [Ngăn xếp: 0]

# <INDENT> <if> <bar> <:>  [Ngăn xếp: 0, 4]

# <INDENT> <x> <=> <42>  [Ngăn xếp: 0, 4, 8]

# <DEDENT> <DEDENT> <else> <:>  [Ngăn xếp: 0]

# <INDENT> <print> <foo>  [Ngăn xếp: 0, 2]

# <DEDENT>
```

Trình phân tích cú pháp (parser) sau đó sẽ xử lý các thẻ "INDENT" và "DEDENT" giống như các dấu phân cách khối (block delimiters).

```python
a = 7
if a > 5:
    print("foo")
else:
    print("bar")
  print("done")
```

```python
if True:
print "true"
```
