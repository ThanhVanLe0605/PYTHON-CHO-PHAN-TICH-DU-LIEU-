Giá trị"True"cónghĩa là giá trị là giá trị bị thiếu trong khi "False"có nghĩa là giá trị không phải là giá trị bị thiếu

Như đã đề cập ở trên, "True"đại diện cho một giá trị bị thiếu và "False"có nghĩalà giá trị có trong tập dữ liệu. Do đó, trong phần thân của vòng lặp for, chúng ta sẽ sử dụng phươngthức .value_counts() để đếm số lượng giá trị "True".

### b. Xử lý giá trị thiếu

Các phương pháp xử lý giá trị bị thiếu:
•Xoá dữ liệu
–Xoá cả dòng
–Xoá cả cột

•Thay thế dữ liệu
–Thay thế bởi giá trị trung bình (mean).
–Thay thế bởi giá trị trung vị (median).
–Thay thế bởi giá trị xuất hiện nhiều nhất (mode).

Lưu ý: Chúng ta chỉ thực hiện phương pháp xoá cả cột khi mà hầu hết các ô giá trị trong cột đềubị trống. Trong tập dữ liệu hiện tại của chúng ta, không có cột nào mà có các giá trị hầu hết là trống.

Ở đây, chúng ta minh hoạt áp dụng phương pháp xoá cho cột "Tin-dung", phương pháp thay thếdùng giá trị trung bình cho cột "Thu-nhap"(numeric feature) và thay thế dùng giá trị xuất hiện nhiềunhất cho cột "Viec-lam"(categorical feature)

**Xóa dữ liệu**

Trong trường hợp, nếu mục tiêu của bài toán hiện tại là xem xét các yếu tố ảnh hưởng đến điểm tíndụng của khách hàng hoặc trong trường hợp cần dự đoán điểm tín dụng của một khách hàng, thì giátrị của cột "Tin-dung"được xem là biến mục tiêu (target variable), đóng vai trò quan trọng. Do đó,khi giá trị của cột "Tin-dung"bị thiếu, chúng ta nên xoá những đó đi. (Nếu sử dụng phương phápthay thế giá trị (bằng mean, median hay mode), thì khả năng cao nó sẽ làm giảm hiệu quả các mô hình dự đoán)

```python
## xóa dòng dữ liệu mà giá trị tại cột Tin_dung là NaN
df.dropna(subset=["Tin-dung"], axis =0, inplace = True)
```

```python
df
```

**Thay thế dữ liệu**

**Thay thế bằng trung bình**

Lưu ý, hiện tại các giá trị trong cột "Thu-nhap"vẫn đang là kiểu "object". Do đó, chưa thể tính giá trịtrung bình trực tiếp mà cần chuyển kiểu "float"rồi mới tính giá trị trung bình (mean). Sau khi tính được giá trị trung bình, chúng ta thay thế giá trị NaN bằng cách sử dụng phương thức $fillna()$ trong Pandas.

```python
x = df['Thu-nhap'].astype("float").mean()
df["Thu-nhap"].fillna(x, inplace = True)
```

**Thay thế bằng giá trị xuất hiện nhiều nhất**

Trong trường hợp cột "Viec-lam", chúng ta thấy phần lớn các khách hàng đều đang có công việc. Dođó, nếu thay thế giá trị bị thiếu NaN bằng giá trị "Dang lam viec"thì khả năng cao là phù hợp vớithực tế. Cụ thể:

```python
df['Viec-lam'].value_counts()
```

Để đảm bảo tính tổng quát, chúng ta sẽ thực hiện thay thế giá trị NaN của cột "Viec-lam"bằngcách sử dụng giá trị xuất hiện nhiều nhất (mode).

```python
x = df["Viec-lam"].mode()[0]
df['Viec-lam'].fillna(x, inplace = True)
```

```python
df
```

Trong trường hợp này, giá trị 380 trở thành giá trị ngoại lệ (outlier), dẫn đến sự sai lệch trong kếtquả phân tích và giảm hiệu quả của các mô hình được xây dựng nếu như không được xử lý

#### b. Xử lý giá trị không phù hợp

Một cách để xử lý các giá trị bị sai là thay thế chúng bằng một giá trị khác.Trong tập dữ liệu ví dụ hiện tại, ở dòng 10 (tương ứng Khách hàng có "Mã KH- 11), rất có thể đây là lỗi đánh máy và giá trị phải là "38"thay vì "380" và chúng ta chỉ cần thay thế giá trị "380" bởi 38

**Thay thế giá trị**

```python
df.loc[10, "Tuoi"] = 38
df
```

Đối với các tập dữ liệu nhỏ, chúng ta có thể thay thế từng dữ liệu sai, nhưng đối với các tập dữ liệu lớn thì điều này thực sự không khả thi

Để thay thế dữ liệu sai cho các tập dữ liệu lớn hơn, chúng ta có thể thiết lập một số quy tắc, vídụ: đặt một số ranh giới cho các giá trị và thay thế bất kỳ giá trị nào nằm ngoài ranh giới

Ví dụ, chúng ta muốn thay thế các giá trị tuổi lớn hơn 100 bằng giá trị 100 và nhỏ hơn 18 bằng giá trị 18




```python
for x in df.index:
  if df.loc[x, "Tuoi"]  > 100:
    df.loc[x, "Tuoi"] = 100
  elif df.loc[x, "Tuoi"] < 18:
    df.loc[x, "Tuoi"] = 18
```

**Thay thế**

Một cách khác để xử lý dữ liệu sai là loại bỏ các dòng chứa dữ liệu sai. Bằng cách này, chúng takhông cần phải tìm ra giá trị để thay thế chúng.

Ví dụ, chúng ta muốn xoá đi các dòng mà giá trị tuổi lớn hơn 100 hoặc nhỏ hơn 18.

```python
for x in df.index:
  if df.loc[x, "Tuoi"] < 18 or df.loc[x, "Tuoi"] > 100:
    df.drop(x, inplace = True)
```

Lưu ý: Phương pháp thay thế và loại bỏ khi xử lý dữ liệu bị sai (wrong data) đều có những ưuđiểm và hạn chế riêng

**Phương pháp thay thế**

Ưu điểm:

- Giữ nguyên kích thước tập dữ liệu: Phương pháp thay thế cho phép giữ nguyênkích thước tập dữ liệu ban đầu, nghĩa là không có mất mát dữ liệu

- Bảo toàn thông tin: Thay vì loại bỏ dữ liệu sai, phương pháp thay thế giữ lạithông tin và có thể cải thiện tính đại diện của tập dữ liệu

Hạn chế
- Khó xác định giá trị thay thế: Đôi khi việc xác định giá trị thay thế phù hợp có thể khá khó, đặc biệt khi dữ liệu sai không rõ ràng

**Phương pháp Loại Bỏ**

Ưu điểm:
- Loại bỏ được nhưng dữ liệu sai: giúp làm sạch tập dữ liệu và đảm bảo tính chính xác của phân tích
- Dễ thực hiện và dễ kiểm tra lại

Hạn chế
- Mất mát thông tin: Loại bỏ dữ liệu có thể dẫn đến mất mát thông tin quý báu và làm giảm độ đại diện của tập dữ liệu.
- Sai lệch (bias): Nếu loại bỏ quá nhiều dữ liệu hoặc dữ liệu quan trọng, có thể gây ra sai lệch trong phân tích và dự đoán

Tóm lại, phương pháp thay thế và loại bỏ đều có ưu điểm và hạn chế riêng. Quyết định sử dụngphương pháp nào phụ thuộc vào ngữ cảnh cụ thể, mục tiêu phân tích, và tính độ chính xác của dữliệu. Trong một số trường hợp, có thể kết hợp cả hai phương pháp để đảm bảo tính chính xác và bảotoàn thông tin quý báu

# 7.2. TINH CHỈNH DỮ LIỆU

Trong quá trình phân tích và xây dựng các mô hình dự đoán, dữ liệu thường được thu thập từ nhiềunguồn khác nhau và có định dạng, đặc điểm riêng biệt. Để làm cho dữ liệu này có thể sử dụng hiệuquả trong các mô hình máy học và thống kê, chúng ta cần chuyển đổi dữ liệu từ các định dạng khácnhau về các định dạng hoặc chuẩn phù hợp cho mục tiêu sử dụng. Thông thương dữ liệu có thể được chia thành 2 nhóm:
- Dữ liệu số (nummeric data)
- Dữ liệu phân loại (categorical data)

## 7.2.1. Chuyển đổi giá trị dữ liệu

Trong tập dữ liệu về Khách hàng đang sử dụng, cột "Thu-nhap"được biểu thị bằng đơn vị VND (ViệtNam Đồng). Giả sử chúng ta đang phát triển một ứng dụng ở một nhóm các quốc gia chấp nhậnđơn vị USD (United States Dollar). Chúng ta cần thực hiện một quá trình chuyển đổi dữ liệu (datatransformation) để chuyển đổi từ VND sang USD.

Giả sử công thức chuyển đổi đơn vị là:

> 1 USD = 24.000 VND



Chuyển đổi giá trị cột thu nhập từ VND sang USD

```python
import pandas as pd
df = pd.read_csv("clean_khach_hang.csv")

df["Thu-nhap"] = df["Thu-nhap"] / 24000

df.rename(columns={"Thu-nhap": "Thu-nhap-USD"}, inplace = True)

df.head()
```

## 7.2.2. Chuyển đổi dữ liệu phân loại sang dữ liệu số



Chuyển đổi dữ liệu sang dạng số là một phần quan trọng trong quá trình tinh chỉnh dữ liệu và xử lýdữ liệu trong khoa học dữ liệu và học máy. Điều này là cần thiết bởi vì nhiều mô hình học máy yêucầu dữ liệu đầu vào dưới dạng số thay vì dữ liệu dạng phân loại. Bên cạnh đó, khi dữ liệu chuyển đổidữ liệu sang dạng số, chúng ta có thể cải thiện hiệu suất của mô hình vì nó có khả năng xử lý cácbiến số tốt hơn, hiểu các mối quan hệ số hơn và thực hiện các phép tính toán dễ dàng hơn.

Dưới đây là một số các kỹ thuật phổ biến được sử dụng để chuyển đổi dữ liệu từ dạng phân loạisang dạng số
- Label Encoding: Mã hóa biến phân loại thành các giá trị số tương ứng. Ví dụ, biến "Goi-tinh"có thể được mã hóa thành 0 và 1, "Nam"và "Nu"tương ứng
- Ordinal Encoding: Mã hóa biến phân loại dạng có thứ tự. Ví dụ: Ví dụ, các giá trị "Daihoc,""Cao Dang"và "Trung hoc"có thể được thay thế thành các giá trị số phản ánh thứ tự này("Dai hoc": 2, "Cao dang": 1, "Trung hoc": 0)
- One-Hot Encoding: Tạo các cột giá trị nhị phân (0 hoặc 1) cho từng giá trị duy nhất trong biếnhạng mục. Điều này giúp mô hình xem xét từng giá trị riêng biệt. Ví dụ, biến "Viec-lam"có thể được chia thành các cột như "Dang lam viec", "That nghiep"và "Sinh vien"

### **Label Encoding**

Ví dụ, chúng ta cần thực hiện mã hóa giới tính sang dạng số 0,1 bằng cách thay các giá trị "Nu"thành"0"và các giá trị "Nam"thành

```python
import pandas as pd
df = pd.read_csv("clean_khach_hang.csv")

df['Gioi-tinh'] = df['Gioi-tinh'].replace({"Nam":0, "Nu":1})

df.head()
```

### **Ordinal Encoding**

Ví dụ, chúng ta cần thực hiện mã hóa cột "Trinh-do"sang dữ liệu số với các giá trị tương ứng là"Trung hoc": 0, "Cao dang": 1, "Dai hoc": 2. Trước tiên, chúng ta cần xây dựng bảng ánh xạ giátrị phân loại vào số tương ứng (ordinal_mapping). Sau đó, thực hiện quá trình ánh xạ dữ liệu bằngphương thức map() trong pandas

```python
ordinal_mapping = {"Trung hoc":0, "Cao dang":1, "Dai hoc": 2}

df['Trinh-do'] = df['Trinh-do'].map(ordinal_mapping)

df.head()
```

### **One-Hot Encoding**

Ví dụ, chúng ta thực hiện mã hóa cột "Viec-lam"có các giá trị "Đang làm việc", "Thất nghiệp", và"Sinh viên"sử dụng phương pháp One-Hot Encoding thông qua phương thức get_dummies() trongpandas để tạo ra các cột giả tương ứng các nhóm giá trị của biến phân loại

```python
df = pd.get_dummies(df, columns=['Viec-lam'])

df.head()
```

## 7.2.3. Chuyển đổi dữ liệu số về dạng phân loại

Chuyển dữ liệu số sang dạng phân loại (categorical data) là một phần quan trọng của qúa trình tinhchỉnh dữ liệu cho phân tích dữ liệu và một số các mô hình máy học. Quá trình này cũng được gọi làphân giỏ dữ liệu (data bining).

Một số trường hợp cần chuyển đổi dữ liệu số sang dạng phân loại:

Mục tiêu của phân tích: Đôi khi, mục tiêu của phân tích yêu cầu dự đoán phân lớp một kháchhàng. Ví dụ, dự đoán rủi ro tính dụng của một khách hàng ("cao", "Trung bình", "Thấp").

•Trực quan hóa dữ liệu: Dữ liệu dạng phân loại thường dễ trực quan hóa hơn. Bạn có thể sửdụng biểu đồ cột, biểu đồ tròn, hoặc biểu đồ thanh để hiển thị phân phối của các nhóm. Ví dụ,gom nhóm các khách hàng theo thu nhập ("Cao", "Trung bình", "Thấp").

•Giảm ảnh hưởng của giá trị cụ thể: Khi dữ liệu số được chuyển sang thành dạng phân loại ởcác nhóm, có thể giảm bớt ảnh hưởng của giá trị cụ thể (outliers) và giúp tạo ra các biểu đồ dễđọc hơn.

Ví dụ, chúng ta muốn phân giá trị của cột "Thu-nhap"thành các nhóm giá trị "Thấp", "Trungbình", và "Cao". Ở đây, chúng ta có 3 nhóm ("Thấp", "Trung bình", "Cao") (hoặc có thể gọi là 3giỏ), do đó cần 4 mốc gía trị để thực hiện quá trình phân chia. Ví dụ chúng ta chọn giá trị thấp nhấtcho cột thu nhập là 0, lớn nhất là vô cùng (float("inf")), và hai mốc nữa 5000000 và 7000000

Công thức phân chia như sau:
> Thấp: thu nhâp từ 0 đến 5000000

> Trung bình: thu nhâp từ 5000000 đến 7000000

> Cao: thu nhập trên 7000000

Trước tiên, chúng ta tạo ra các giỏ (bins) bằng cách định nghĩa khoảng giá trị cho từng nhóm.Tiếp theo, thiết lập nhãn (labels) cho từng nhóm dữ liệu. Cuối cùng, sử dụng phương thức cut() trong pandas để thực hiện phân nhóm cột "Thu-nhap

```python
import  pandas  as pd

df = pd.read_csv("clean_khach_hang.csv")

bins = [0, 5000000, 7000000, float("inf")]

labels = ["Thap", "Trung binh", "Cao"]

df["Thu-nhap"] = pd.cut(df['Thu-nhap'], bins = bins,
                        labels = labels, right = False)
```
