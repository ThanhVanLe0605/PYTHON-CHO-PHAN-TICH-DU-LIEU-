Các kiểu dữ liệu (data types) thực chất là các biến (variables) mà bạn sử dụng để dành sẵn một vùng không gian trong bộ nhớ. Các biến trong Python không cần khai báo tường minh (explicit declaration) để cấp phát không gian bộ nhớ. Quá trình khai báo diễn ra tự động khi bạn gán một giá trị cho biến.

# Section 2.1: String Data Type

Chuỗi được nhận diện là một tập hợp các ký tự liên tiếp được đặt trong dấu ngoặc kép. Python cho phép sử dụng các cặp dấu ngoặc đơn hoặc ngoặc kép. Chuỗi là kiểu dữ liệu tuần tự bất biến (immutable sequence data type), nghĩa là mỗi khi bạn thực hiện bất kỳ thay đổi nào trên chuỗi, một đối tượng chuỗi hoàn toàn mới sẽ được tạo ra.

```python
a_str = 'Hello World'
print(a_str)
print(a_str[0])
print(a_str[0:5])
```

# Section 2.2: Set Data Types

Tập hợp (Set) là các bộ sưu tập không có thứ tự (unordered collections) gồm các đối tượng duy nhất (unique objects). Có hai loại tập hợp:

  * **Sets (Tập hợp thông thường)** - Chúng có tính khả biến (mutable) và các phần tử mới có thể được thêm vào sau khi tập hợp đã được định nghĩa.

```python
basket = {'a','a','b', 'b', 'c', 'd','d', 'e', 'f'}
print(basket) # Các phần tử trùng lặp sẽ bị loại bỏ
```

```python
tap_hop = set('abracadabra')
print(tap_hop) # các chữ cái duy nhất trong biến tap_hop
```

```python
tap_hop.add('z')
print(tap_hop)
```

  * **Frozen Sets (Tập hợp đóng băng)**- Chúng có tính bất biến (immutable) và các phần tử mới không thể được thêm vào sau khi đã định nghĩa.

```python
dong_bang = frozenset('abracadabra')
print(dong_bang)
```

```python
cities = frozenset(['TP HCM', 'Thủ Đức', 'Hà Nội'])
print(cities)
```

```python
cities.add(['Hải Phòng']) # Lỗi do frozenset là immutable - không thực hiện chỉnh sửa trên nó
```

# Section 2.4: List Data Type

Một danh sách chứa các phần tử được phân tách bằng dấu phẩy và đặt trong dấu ngoặc vuông []. Danh sách gần như tương tự với mảng (arrays) trong ngôn ngữ C. Điểm khác biệt là tất cả các phần tử thuộc về một danh sách có thể có các kiểu dữ liệu khác nhau.

```python
list = [123, 'abcd', 10.2, 'd'] # Có thể là mảng chứa nhiều kiểu dữ liệu hoặc một kiểu dữ liệu duy nhất.
list1 = ['hello', 'world']
print(list)               # Sẽ xuất ra toàn bộ danh sách. [123,'abcd',10.2,'d']
print(list[0:2])          # Sẽ xuất ra hai phần tử đầu tiên của danh sách. [123,'abcd']
print(list1 * 2)          # Sẽ nhân đôi list1. ['hello','world','hello','world']
print(list + list1)       # # Sẽ là phép nối (concatenation) của cả hai danh sách. [123,'abcd',10.2,'d','hello','world']
```

# Section 2.5: Dictionary Data Type

Từ điển bao gồm các cặp khóa-giá trị (key-value pairs). Nó được đặt trong dấu ngoặc nhọn {} và các giá trị có thể được gán cũng như truy xuất bằng dấu ngoặc vuông [].

```python
dict = {'name': 'red',
        'age' :  10   }
print(dict)             # Sẽ xuất ra tất cả các cặp key-value. {'name':'red','age':10}
print(dict['name'])     # Sẽ xuất ra chỉ giá trị gắn với khóa 'name'. 'red'
print(dict.values())    # Sẽ xuất ra danh sách các giá trị trong dic. ['red',10]
print(dict.keys())      # Sẽ xuất ra danh sách các khóa. ['name','age']
```

# Section 2.6: Tuple Data Type

Danh sách (List) được đặt trong dấu ngoặc vuông [] và các phần tử cũng như kích thước của chúng có thể bị thay đổi, trong khi tuples được đặt trong dấu ngoặc đơn () và không thể cập nhật. Tuples là bất biến (immutable).

```python
tuple_ = (123, 'hello')
tuple1 = ('world',)
# tuple1 = ('world') -> là chuỗi, cần có dấu phẩy để hiểu là tuple. Khi thực hiện nối tuple mới không bị lỗi
print(tuple_)                 # Sẽ xuất ra toàn bộ tuple. (123,'hello')
print(tuple_[0])              # Sẽ xuất ra giá trị đầu tiên. (123)
```

```python
print(tuple_ + tuple1)        # Sẽ xuất ra (123,'hello','world')
```

```python
tuple_[1] = 'update'          # Lệnh này sẽ báo lỗi (error) vì tuple không thể thay đổi
```
