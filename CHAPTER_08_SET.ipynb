    * **Các phép toán xử lý không gian dữ liệu chính:**

**Các phép kiểm tra (Validation Checks):**

  * Truy vấn (in): Kiểm tra nhanh sự tồn tại của một điểm dữ liệu cụ thể.

**Đo lường (Measurement):**

Dùng hàm len() để trích xuất số lượng phần tử độc nhất của tập dữ liệu

# 8.1.Các phép toán trên tập hợp dữ liệu (Operations on sets)

Tương tác giữa các tập dữ liệu (với các set khác):

## 8.1.1. Phép giao (intersection)


```python
print({1, 2, 3, 4, 5}.intersection({3, 4, 5, 6}))
print({1, 2, 3, 4, 5} & {3, 4, 5, 6})
```

##  8.1.2. Phép hợp (union)

```python
# Gộp các tập dữ liệu (tự động loại bỏ trùng lặp)
print({1, 2, 3, 4, 5}.union({3, 4, 5, 6}))
print({1, 2, 3, 4, 5} | {3, 4, 5, 6})
```

## 8.1.3. Phép hiệu (difference)

```python
# Lọc bỏ các điểm dữ liệu đã tồn tại ở tập tham chiếu
print({1, 2, 3, 4}.difference({2, 3, 5}))
print({1, 2, 3, 4} - {2, 3, 5})
```

## 8.1.4. Phép hiệu đối xứng (symmetric difference)

```python
# Giữ lại các điểm dữ liệu dị biệt của cả hai tập
print({1, 2, 3, 4}.symmetric_difference({2, 3, 5}))
print({1, 2, 3, 4} ^ {2, 3, 5})
```

## 8.1.5. Kiểm tra tập cha (superset check)

```python
# Kiểm tra tính bao hàm toàn bộ dữ liệu
print({1, 2}.issuperset({1, 2, 3}))
print({1, 2} >= {1, 2, 3})
```

## 8.1.6. Kiểm tra tập con (subset check)

```python
print({1, 2}.issubset({1, 2, 3}))
print({1, 2} <= {1, 2, 3})
```

## 8.1.7. Kiểm tra tính rời rạc

```python
print({1, 2}.isdisjoint({3, 4}))
print({1, 3}.isdisjoint({3, 4}))
```

**Tương tác với các điểm dữ liệu đơn lẻ**

```python
# Kiểm tra sự tồn tại
print(2 in {1, 2, 3})
print(4 in {1, 2, 3})
print(4 not in {1, 2, 3})
```

# 8.2. Trích xuất các phần tử độc nhất từ một danh sách (Deduplication)

```python
restaurants = ["McDonald's", "Burger King", "McDonald's", "Chicken Chicken"]
unique_restaurants = set(restaurants)
print(unique_restaurants)
```

```python
list(unique_restaurants)
```

```python
# Khử toàn bộ dữ liệu trùng lặp và trả về một list mới
list(set(restaurants))
```
