* Con số -18000 đại diện cho độ lệch (offset) so với Giờ Quốc tế Phối hợp (UTC), được tính bằng giây.

# 5.2. Khởi tạo các đối tượng datetime có nhận thức múi giờ

Theo mặc định, tất cả các đối tượng datetime trong Python đều là dạng "naive" (ngây ngô - không có thông tin về múi giờ). Để biến chúng thành đối tượng "aware" (có nhận thức múi giờ), bạn phải gắn thêm một đối tượng tzinfo. Đối tượng này cung cấp độ lệch so với giờ UTC (UTC offset) và tên viết tắt của múi giờ dựa trên ngày và giờ cụ thể.

## 5.2.1. Múi giờ có độ lệch cố định (Fixed Offset Time Zones)

Đối với các múi giờ có độ lệch cố định so với UTC, từ phiên bản Python 3.2 trở đi, module datetime cung cấp lớp timezone. Đây là một triển khai cụ thể của tzinfo, nhận vào một tham số timedelta và một tham số tên gọi (không bắt buộc

**Đối với quốc gia KHÔNG có DST (Như Việt Nam, Nhật Bản)**

Vì múi giờ là cố định, chỉ cần dùng lớp timezone có sẵn của Python để gắn thẳng vào lúc tạo thời gian.

```python
from datetime import datetime, timedelta, timezone

# Khởi tạo múi giờ JST(UTC+9)
JST = timezone(timedelta(hours=+9))

dt = datetime(2015, 1, 1, 12, 0, 0, tzinfo = JST)
print(dt)
# KQ: 2015-01-01 12:00:00+09:00

print(dt.tzname())
# KQ: UTC+09:00

# Khởi tạo với tên cụ thể
dt = datetime(2015, 1, 1, 12, 0, 0, tzinfo = timezone(timedelta(hours=9), 'JST'))
print(dt.tzname())
# KQ: JST
```

## 5.2.2. Các múi giờ có Giờ tiết kiệm ánh sáng ngày (Daylight Saving Time - DST)

**Đối với quốc gia CÓ DST**

Thư viện chuẩn của Python không cung cấp sẵn lớp cho các múi giờ có DST, vì vậy cần sử dụng thư viện pytz hoặc dateutil.

**Sử dụng dateutil:**
Có thể dùng phương thức tz.gettz() để lấy đối tượng múi giờ và truyền trực tiếp vào hàm khởi tạo datetime

```python
from datetime import datetime
from dateutil import tz

# 'Asia/Ho_Chi_Minh'
local = tz.gettz('Asia/Ho_Chi_Minh') # Múi giờ địa phương
# Hãy tự động lấy múi giờ hiện tại đang được cài đặt trên cái máy tính/server chạy đoạn code này
dt_hien_tai = datetime.now(tz = local)
print(dt_hien_tai)

do_lech = dt_hien_tai.strftime('%z')
print("Chỉ lấy độ lệch:")
print(do_lech)
```

```python
# Múi giờ Thái Bình dương (Mỹ)
PT = tz.gettz('US/Pacific')

dt_pst = datetime(2015, 1, 1, 12, tzinfo = PT)
dt_pdt = datetime(2015, 7, 1, 12, tzinfo =PT) # DST được tự động xử lý

print(dt_pst) # Giờ tiêu chuẩn (standard time)
# KQ: 2015-01-01 12:00:00-08:00
print(dt_pdt) # Giờ tiết kiệm ánh sáng (Daylaight Time)
# KQ: 2015-07-01 12:00:00-07:00
```

Múi giờ thực chất không chỉ là một con số cố định (như +7 hay -8), mà nó là một tập hợp các quy tắc của một khu vực địa lý. Khu vực US/Pacific (Bờ Tây nước Mỹ) có quy tắc như sau:

* Vào mùa đông (tháng 1): Họ dùng giờ tiêu chuẩn (Pacific Standard Time - PST). Độ lệch là UTC-8.

* Vào mùa hè (tháng 7): Để tận dụng ánh sáng mặt trời, họ vặn đồng hồ nhanh lên 1 tiếng. Lúc này họ chuyển sang giờ mùa hè (Pacific Daylight Time - PDT). Độ lệch bị rút ngắn lại thành UTC-7.

* Khi dùng thư viện dateutil (hoặc pytz), đối tượng PT (US/Pacific) cực kỳ thông minh. Nó tự động nhìn vào cái ngày truyền vào:

* Thấy (2015, 1, 1) -> À, mùa đông -> Tự động gắn mác -08:00.

* Thấy (2015, 7, 1) -> À, mùa hè -> Tự động gắn mác -07:00.

Đó là lý do vì sao chỉ dùng đúng một biến PT nhưng kết quả in ra lại có 2 độ lệch UTC khác nhau. Mùa (thời điểm trong năm) trực tiếp quyết định độ lệch giờ của khu vực đó.

**Sử dụng pytz:**

Tất cả các trường hợp biên đều được xử lý đúng khi dùng pytz. Tuy nhiên, không nên gắn trực tiếp múi giờ pytz qua tham số tzinfo của hàm khởi tạo. Thay vào đó, hãy sử dụng phương thức localize() của đối tượng múi giờ

```python
from datetime import datetime, timedelta
import pytz

PT = pytz.timezone('US/Pacific')

# Sử dụng localize thay vì truyền tzinfo vào datetime
dt_pst = PT.localize(datetime(2015, 1, 1, 12))
dt_pdt = PT.localize(datetime(2015, 11, 1, 0, 30))

print(dt_pst)
print(dt_pdt)
```

*Lưu ý quan trọng khi tính toán với pytz:*

Nếu bạn thực hiện phép tính cộng/trừ thời gian (timedelta) trên đối tượng datetime của pytz, bạn phải gọi phương thức normalize() để cập nhật lại thông số múi giờ nếu phép tính đó vượt qua ranh giới thay đổi DST:

```python
dt_new = dt_pdt + timedelta(hours=3)
print(dt_new)

dt_corrected = PT.normalize(dt_new)
print(dt_corrected)
```

# 5.3. Tính toán chênh lệch thời gian

Đối tượng timedelta (nằm trong module datetime) cực kỳ hữu ích để tính toán sự chênh lệch giữa các mốc thời gian.

## 5.3.1. Tính khoảng cách giữa hai mốc thời gian (Tính Delta)

Khi bạn lấy hai đối tượng datetime trừ cho nhau, kết quả trả về không phải là một con số thông thường, mà là một đối tượng thuộc kiểu timedelta. Đối tượng này lưu trữ chính xác khoảng thời gian chênh lệch.

```python
from datetime import datetime, timedelta
import pytz
from dateutil import tz

hcm = tz.gettz('Asia/Ho_Chi_Minh') # Lấy thời gian hiện tại
now = datetime.now(tz = hcm)

then = datetime(2026, 4, 17, tzinfo=hcm)
# Tạo mốc thười gian quá khứ và gán múi giờ
# Lưu ý: Việc truyền giờ/phút/giây là không bắt buộc khi tạo đối tượng datetime mới.
# Nếu không truyền, nó sẽ mặc định là 0 giờ 0 phút 0 giây.

delta_created = now - then
# Biến 'delta' lúc này mang kiểu dữ liệu là timedelta
print(delta_created.days)

print(delta_created.seconds)
```

Thuộc tính .seconds chỉ trả về số giây lẻ trong một ngày (tối đa 86399). Nếu bạn muốn biết tổng cộng khoảng thời gian đó dài bao nhiêu giây, bạn phải dùng hàm .total_seconds().

## 5.3.2. Cộng/Trừ thời gian (Lấy N ngày trước hoặc sau)

Thay vì trừ hai ngày cho nhau để ra timedelta, bạn có thể tạo ra một timedelta (ví dụ: 120 ngày) rồi cộng hoặc trừ nó vào thời gian hiện tại để tính ra mốc thời gian trong tương lai hoặc quá khứ.

**Hàm lấy ngày tương lai (sau N ngày)**

```python
def get_n_days_after_date(date_format = "%d %B %Y", add_days = 120):
  # Lấy giờ hiện tại CỘNG thêm 120 ngày
  date_n_days_after = datetime.now() + timedelta(days= add_days)

  # Định dạng lại kết quả thành chuỗi dễ đọc
  return date_n_days_after.strftime(date_format)
```

**Hàm lấy ngày quá khứ**

```python
def get_n_days_before_date(self, date_format ="%d %B %Y", days_before = 120):
  # Lấy giờ hiện tại TRỪ đi 120 ngày
  date_n_days_ago = datetime.now() - timedelta(days = days_before)

  return date_n_days_ago.strftime(date_format)
```

```python
# Thử lấy biến noon (kiểu time) trừ đi biến today (kiểu date)
noon - today
```

**Cách giải quyết: Đồng bộ kiểu dữ liệuliệu**

Tuy nghiêm ngặt, nhưng Python cung cấp cách rất dễ dàng để chuyển đổi và kết hợp các kiểu dữ liệu này lại với nhau. Nếu  muốn tính khoảng cách giữa một ngày (date) và một khoảnh khắc trọn vẹn (datetime),  phải nâng cấp biến date đó lên thành datetime.

Có 2 cách phổ biến để làm việc này:

```python
# Cách 1: Xây lại từ đầu bằng cách trích xuất thuộc tính (.year, .month, .day)
# Tạo một datetime mới từ năm/tháng/ngày của biến `today`, giờ phút giây mặc định là 0.
today_midnight = datetime.datetime(today.year, today.month, today.day)
millenium_turn = datetime.datetime(2000, 1, 1, 0, 0, 0) #Khoảng khắc giao thừa năm
print('Thời gian từ năm 2000 đến nửa đêm hôm nay: ', today_midnight - millenium_turn)
```

```python
# Cách 2: Dùng .combine()
```

Hàm .combine() lấy một đối tượng date ghép với một đối tượng time để đẻ ra một đối tượng datetime hoàn chỉnh

```python
# Ghép `today` (Ngày hôm nay) với `noon` (12 giờ trưa) thành -> Trưa hôm nay
today_noon = datetime.datetime.combine(today, noon)

print('Thời gian từ năm 2000 đến 12h trưa hôm nay: ', today_noon - millenium_turn)
```

# 5.5. Chuyển đổi qua lại giữa các múi giờ

Để chuyển đổi giữa các múi giờ, đối tượng datetime  BẮT BUỘC phải là timezone-aware (có nhận thức múi giờ)

```python
from datetime import datetime
from dateutil import tz

# Bước 1: Khởi tạo múi giờ theo utc, và theo địa phương
utc = tz.tzutc()
local = tz.tzlocal()

# Bước 2: Lấy thời gian hiện tại theo mốc UTC
utc_now = datetime.utcnow()
# LƯU Ý: Hàm utcnow() lấy đúng giờ UTC, nhưng nó lại trả về một đối tượng "naive" (ngây ngô).
# Tức là nó mang giá trị của giờ UTC, nhưng bên trong nó không hề dán cái nhãn "Tôi là UTC".

# Bước 3: Dán nhãn múi giờ cho utc_now
utc_now = utc_now.replace(tzinfo = utc)

# Bước 4: Chuyển đổi múi giờ
local_now = utc_now.astimezone(local)
local_now
# Hàm .astimezone() là "trái tim" của phép chuyển đổi. Nó nhìn vào thời gian UTC đang có,
# tính toán độ lệch, và dịch chuyển các con số để khớp với múi giờ local.
```

# 5.6. Các phép tính ngày tháng cơ bản

Trong thực tế, chúng ta hiếm khi chỉ khởi tạo một ngày rồi để đó. Bạn sẽ thường xuyên phải tìm khoảng cách giữa các ngày, hoặc tính xem ngày mai/ngày hôm qua là ngày mấy. Điều này được xử lý gọn gàng bằng đối tượng timedelta

```python
import datetime

# Lấy ngày hôm nay
today = datetime.date.today()
print("Hôm nay:", today)

# Lấy hôm nay TRỪ đi một ra ngày hôm qua
yesterday = today - datetime.timedelta(days = 1)
print('Hôm qua:', yesterday)

# Lấy hôm nay CỘNG thêm một ngày ra ngày mai
tomorrow = today + timedelta(days = 1)
print('Ngày mai:', tomorrow)

# Lất ngày mai trừ đi hôm qua
print('Khoảng cách giữa ngày mai và hôm qua:', tomorrow - yesterday)

```

**Điểm hay của timedelta:**
* Không cần bận tâm về việc qua tháng hay qua năm nhuận. Nếu hôm nay là 2024-03-01 (Năm nhuận), khi trừ đi 1 ngày, Python tự động biết lùi về 2024-02-29 chứ không phải mùng 1 hay 28.

# 5.7. Chuyển đổi Timestamp sang Datetime

Module datetime có thể chuyển một POSIX timestamp thành một đối tượng datetime chuẩn UTC.
(Lưu ý: Mốc thời gian Epoch - Kỷ nguyên máy tính - được quy ước là Khoảnh khắc nửa đêm Ngày 1 Tháng 1 Năm 1970).

**Timestamp là gì?**
* Máy tính không lưu thời gian dưới dạng "Ngày 22 Tháng 7". Nó đếm tổng số giây đã trôi qua kể từ ngày 1/1/1970. Cái con số khổng lồ đó gọi là Timestamp.

```python
from datetime import timezone
# Lấy timestamp hiện tại (số giây từ năm 1970)
seconds_since_epoch = time.time()

# Dịch timestamp và dán luôn nhãn UTC vào
chuan_utc = datetime.fromtimestamp(seconds_since_epoch, tz=timezone.utc)
chuan_utc
```

# 5.8. Cách trừ đi số tháng một cách chính xác(*)

* timedelta không có tham số months , years

Lý do rất thực tế: Một ngày luôn có 24 giờ, nhưng một tháng lại có thể có 28, 29, 30 hoặc 31 ngày. Nếu ngày hiện tại là 31/03/2013, trừ đi 1 tháng thì ra ngày mấy? Tháng 2 làm gì có ngày 31?

**Cách 1: Tự viết hàm thủ công bằng toán học (Dùng module calendar)**

```python
import calendar
from datetime import date

def monthdelta(date, delta):
  # Tính tháng và năm mới
  m, y = (date.month + delta) % 12 , date.year + ((date.month) + delta - 1) //12
  if not m:m =12

  d = min(date.day, calendar.monthrange(y, m)[1])

  return date.replace(day =d, month =m, year = y)

next_month = monthdelta(date.today(), 1)
next_month
```

**Cách 2: Dùng thư viện dateutil**

```python
import datetime
import dateutil.relativedelta

# Lấy mốc ngày 31 tháng 3 năm 2013
d = datetime.datetime.strptime("2026-04-18", "%Y-%m-%d")

# Dùng relativedelta để cộng thêm đúng 1 tháng
d2 = d + dateutil.relativedelta.relativedelta(months = 1)

print(d2)
```

Điều này sinh ra 2 rắc rối lớn:

# 5.11. Chuyển đổi chuỗi có chứa tên múi giờ viết tắt thành đối tượng Datetime (Timezone-aware)

**Vấn đề cốt lõi:**
* Đôi khi, hệ thống nhận được những chuỗi thời gian chứa tên múi giờ viết tắt, ví dụ: '2014-01-02 04:00:00 EST' hoặc 'CST'

* Sự thật phũ phàng là các tên viết tắt này không hề độc nhất (ambiguous).

VD: CST có thể là Central Standard Time (Mỹ), China Standard Time (Trung Quốc), hoặc Cuba Standard Time (Cuba).

Vì máy tính không thể tự "đoán" được đang nói đến nước nào,  bắt buộc phải tự định nghĩa một "cuốn từ điển" (mapping) để giải thích các từ viết tắt này cho Python hiểu.

**Cách giải quyết bằng thư viện dateutil**

Dùng hàm parse của thư viện dateutil, kết hợp với một dictionary (tzinfos) đóng vai trò là "từ điển dịch thuật".

```python
from dateutil import tz
from dateutil.parser import parse

# 1. Khai báo các đối tượng múi giờ chuẩn của Mỹ
ET = tz.gettz('US/Eastern')
CT = tz.gettz('US/Central')
MT = tz.gettz('US/Mountain')
PT = tz.gettz('US/Pacific')

# 2. Tạo "từ điển dịch thuật"
# Báo cho python biết: nếu thấy chữ 'CST' hoăc 'CDT' thì lấy múi giờ CT (US/Central)
us_tzinfos = {

              'CST': CT, 'CDT':CT,
              'EST':ET,  'EDT':ET,
              'MST':MT, 'MDT':MT,
              'PST':PT, 'PDT':PT
}

# 3. Tiến hành đọc chuỗi (parse) và truyền cuốn từ điển vào
dt_est = parse('2014-01-02 04:00:00 EST', tzinfos = us_tzinfos)
dt_pst = parse('2016-03-11 16:00:00 PST', tzinfos = us_tzinfos)

print(dt_est)
```

# 5.12. Phân tích ngày giờ linh hoạt (Trích xuất ngày giờ từ một đoạn văn bản thô)

Trong thực tế, người dùng hoặc các trang web hiếm khi nhập đúng định dạng chuẩn như 2023-10-25. Họ thường viết thành những câu văn dài dòng như: "Hôm nay là ngày 1 tháng 1 năm 2047 vào lúc 8:21 sáng".

Làm sao để tách cái ngày giờ đó ra khỏi đống chữ nghĩa dư thừa? Chúng ta hoàn toàn có thể trích xuất ngày tháng từ một đoạn văn bản bằng cách sử dụng hàm parse của dateutil ở chế độ "fuzzy" (mờ / linh hoạt). Ở chế độ này, hàm sẽ tự động bỏ qua tất cả những từ ngữ không thuộc về định dạng ngày giờ.

```python
from dateutil.parser import parse

# Bật chế độ fuzzy = True để máy tự động tìm
dt = parse("Today is January 1, 2047 at 8:21:00AM", fuzzy = True)

print(dt)
```

*Cách hàm này hoạt động dưới nền:*

1. Nó nhận vào cả một câu văn dài: "Today is January 1, 2047 at 8:21:00AM"

2. Nhờ bật fuzzy=True, nó đi càn quét và nhận ra các chữ "Today", "is", "at" không liên quan gì đến lịch hay đồng hồ. Trình phân tích sẽ xóa sổ các chữ này.

3. Phần còn lại được giữ lại là: "January 1, 2047 8:21:00AM".

4. Cuối cùng, nó chuyển đổi chuỗi sạch sẽ đó thành đối tượng datetime.

💡 Lưu ý thực chiến (Góc nhìn Data)
Tính năng fuzzy=True cực kỳ xịn, nhưng bạn cần cẩn thận vì nó không phải là Trí tuệ nhân tạo (AI), nó chỉ dựa trên các quy tắc đoán chữ. Do đó, nó có một vài điểm mù:

Văn bản tiếng Anh: Nó hoạt động hoàn hảo với văn bản tiếng Anh (January, Monday, AM, PM...). Tuy nhiên, nếu bạn ném vào câu tiếng Việt "Hôm nay là ngày 15 tháng 8", nó sẽ không hiểu được.

Cẩn thận với các con số lạc loài: Nếu trong câu của bạn có chứa những con số không phải là ngày tháng, nó có thể ghép nhầm.

Ví dụ: Nếu bạn parse câu "I bought 3 apples on Jan 5th", nó có thể cố tình nhặt số 3 kia đem ghép vào ngày hoặc năm, dẫn đến kết quả bị sai lệch hoàn toàn.

```python
parse("Hôm nay là ngày 15 tháng 8", fuzzy = True)
```

# 5.13. Lặp qua một chuỗi các ngày (Iterate over dates)

Có thể lặp qua cái chuoi_ngay này y hệt như một danh sách bình thường:
